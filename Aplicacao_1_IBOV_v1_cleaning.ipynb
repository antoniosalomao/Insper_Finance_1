```python
import pandas as pd
import numpy as np
import openpyxl

ibov_path = ("/Users/antonioelias/Desktop/Insper - Modules/Insper Q3/Finance 1 - Investments/Aplicacao_1/Ibovespa_Consolidado.xlsx")
df_main = pd.read_excel(ibov_path)
df_main.set_index('Date', inplace=True)
df_main.drop(df_main.columns[:2], axis=1, inplace=True)
df_main.tail(10)
```

```python
# Data wrangling:

cols_obj = df_main.columns

cols = []
for i in cols_obj:
    cols.append(i)

new_cols = []
for col_str in cols:
    new_col = col_str[-5:]
    new_cols.append(new_col)

# Setting new column names:
df_main.columns = new_cols
df_main.tail()
```

```python
# Searching for NaN and Null values:
# Searching for '-'

k_list = []

null_dict = {}
for col in new_cols:
    k = df_main[col].str.count('-').sum()
    null_dict[col] = k

# 204 total obeservations
# Lets say we want to save about 80% of our data
# (0.8) * 204 = 163.2 --> 163 (minimum requirement)

# DataFrame cleaner:

def clean_df(df, min_data):

    count = int(len(df))
    alpha = 1 - min_data
    max_null = count*(alpha)

    df_demo = df.copy()
    
    for k, v in null_dict.items():
        if v >= max_null:
            df_demo.drop(k, axis=1, inplace=True)
    
    to_drop = ['-']
    df = df_demo[~df_demo[:].isin(to_drop)]
    df.dropna(axis=0, inplace=True)
    df.reset_index(inplace=True)
    
    return df

df_v1 = clean_df(df_main, 0.97)
df_v1
```
